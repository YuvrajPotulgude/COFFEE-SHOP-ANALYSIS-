```python
df['size']=np.where(df['product_detail'].str.contains('Rg'),'regular',
            np.where(df['product_detail'].str.contains('Lg'),'large',
            np.where(df['product_detail'].str.contains('Sm'),'small','not defined')))
```

```python
df['product_detail']=df['product_detail'].str.replace(r'\b(rg|lg|sm)\b','',case=False,regex=True)
df['product_detail']=df['product_detail'].str.strip()
```

```python
string_columns=df.select_dtypes(include='object').columns
whitespace_columns={col:df[col].astype(str).str.contains(r'\s',regex=True).any() for col in string_columns}
for col ,has_whitespace in whitespace_columns.items():
    print(f"column'{col}' contais  whitespace:{has_whitespace}")
```

```python
df=df.apply(lambda col:col.map(lambda x:x.strip() if isinstance(x,str) else x))
```

```python
string_columns=df.select_dtypes(include='object').columns
whitespace_columns={col:df[col].astype(str).str.contains(r'\s',regex=True).any() for col in string_columns}
for col ,has_whitespace in whitespace_columns.items():
    print(f"column'{col}' contais  whitespace:{has_whitespace}")
```

```python
df['store_location']=df['store_location'].str.strip()
```

```python
df.index=df.index+1
```

```python
df['Total_Amount']=df['transaction_qty']*df['unit_price']
```

```python
df.head()
```

```python
df['transaction_date']=pd.to_datetime(df['transaction_date'])
```

```python
df['transaction_day']=df['transaction_date'].dt.day_name()
```

```python
df['transaction_month']=df['transaction_date'].dt.month_name()
```

```python
df['transaction_hour']=df['transaction_time'].apply(lambda x:x.hour)
```

```python
df.head(15)
```

```python
pivot_hour=pd.pivot_table(df,values='transaction_id',index='transaction_hour',aggfunc='count')
print(pivot_hour)
```

```python
def format_currency(value):
    return"${:,.2f}".format(value) 
```

```python
pivot_weekday=pd.pivot_table(df,values='transaction_id',index='transaction_day',aggfunc='count')
print(pivot_weekday)
```

```python
pivot_month=pd.pivot_table(df,values='Total_Amount',index='transaction_month',aggfunc='sum')
month_order=['January','Feburary','March','April','May','June','July','August','September','October','November','December']
pivot_month=pivot_month.reindex(month_order)
print(pivot_month)
```

```python
pivot_product=pd.pivot_table(df,values='Total_Amount',index='product_category',aggfunc='sum')
print(pivot_product)
total_sum=pivot_product['Total_Amount'].sum()
pivot_product['total_sum']=total_sum
print(f"total_profit=       {  total_sum}")
```

```python
pivot_product_type=pd.pivot_table(df,values='Total_Amount',index='product_type',aggfunc='sum')
top_5=pivot_product_type.sort_values(by='Total_Amount',ascending=False).head(5)
print(top_5)
total=pivot_product['Total_Amount'].sum()
print(f"total_Amount=         {total}")
```

```python
pivot_store_location=pd.pivot_table(df,values='Total_Amount',index='store_location',aggfunc='sum').sort_values(by='Total_Amount',ascending=False)
print(pivot_store_location)
total=pivot_store_location['Total_Amount'].sum()
print(f"total_Amount=         {total}")
```

```python
pivot_cstore_location=pd.pivot_table(df,values='transaction_id',index='store_location',aggfunc='count')
print(pivot_cstore_location)
```

```python
pivot_size=pd.pivot_table(df,values='transaction_id',index='size',aggfunc='count')
print(pivot_size)
```

```python
pivot_product_detail=pd.pivot_table(df,values='Total_Amount',index='product_detail',aggfunc='sum').sort_values(by='Total_Amount',ascending=False).head(5)
print(pivot_product_detail)
```

```python
import matplotlib.pyplot as plt
ax=pivot_hour.plot(kind='line',marker='o',color='#38220f')
ax.set_facecolor('#dbc1ac') 
plt.gcf().set_facecolor('#dbc1ac')
plt.grid(True,color='#634832',linewidth=1)
plt.title('hour vs sales ')
plt.xlabel('hour')
plt.ylabel('sales')
plt.grid(True)
plt.show()
```
